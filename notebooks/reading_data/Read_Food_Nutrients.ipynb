# Reading food Nutrient data

```python
import numpy as np
import pandas as pd
import xarray as xr
```

```python
nutrients = pd.read_excel("../../data/food/Nutrition data.xlsx", sheet_name="Match_Items_Nutrition", header=1)
nutrients
```

```python
column_names = nutrients.columns[4:]
column_names
```

```python
functional_weight = nutrients["Default quantity - from NDNS"].astype(str) + " " + nutrients["Units - a tidied up version of the Units given in NDNS (e.g. 'grams' -> 'Grams')"]
functional_weight
```

```python
nutrient_names = []
units = []

for column in column_names:
    try:
        nutrient_names.append(column[:column.index("(", -6)-1])
        units.append(column[column.index("(", -6):])
    except ValueError:
        nutrient_names.append(column)
        units.append(" ")
```

```python
item_ids = nutrients["FAOSTAT ID"]
item_names = nutrients["FAOSTAT balance food name"]

for id, name in zip(item_ids, item_names):
    print(id, name)
```

```python
values = nutrients.iloc[:,4:]
values
```

```python
data = xr.DataArray(
    data=values,
    dims=["Item", "Nutrient"],
    
    coords = dict(
        Item=("Item", item_ids),
        Name=("Item", item_names)
        Weight=("Item", functional_weight),
        Nutrient=("Nutrient", nutrient_names),
        Units=("Nutrient", units),
    )
)
data
```

```python
data.sel(Item=2615, Nutrient="Energy")
```

```python
data.to_netcdf("../../agrifoodpy/food/data/Nutrients.nc")
```
