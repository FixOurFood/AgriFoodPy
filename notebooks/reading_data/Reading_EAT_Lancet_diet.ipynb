# Reading EAT Lancet Diet

```python
import numpy as np
import pandas as pd
import xarray as xr

from agrifoodpy.food.food_supply import plot_bars
```

TO-DO: Explain where data comes from

```python
diet = pd.read_excel("../../data/food/EAT Lancet.xlsx", header=0)[:-2]
diet = diet[diet["Quantity__g"] > 0]
diet
```

```python
item_names = diet["FAOSTAT_balance_Item_name"]
qty = diet["Quantity__g"]
units = np.repeat("g", len(qty))
```

```python
data = xr.DataArray(
    data=qty,
    dims=["Item"],
    coords = dict(
        Item=("Item", item_names),
        Units=("Item", units),
    )
)
data
```

```python
data.to_netcdf("../../agrifoodpy/food/data/EAT_Lancet.nc")
```
